1次元の移流方程式

$$
    \frac{\partial q}{\partial t} + c \frac{\partial q}{\partial x} = 0
$$

離散化

$$
    \frac{q(x, t+\Delta t)-q (x, t)}{\Delta t} + c \frac{q(x+\Delta x, t)-q (x, t)}{\Delta x} = 0
$$

式変形

$$
    q(x, t+\Delta t) = q (x, t) + c \frac{\Delta t}{\Delta x}\left\{q(x, t) - q(x+\Delta x, t)\right\}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def simulation(len_x, delta_x, len_t, delta_t, c, q_init):
    """
    """
    q = np.zeros((len_t, len_x))
    q[0] = q_init
    x = np.arange(0, len_x*delta_x, delta_x)

    for i_t in range(1, len_t):
        q[i_t] = q[i_t-1] + np.pad(c * delta_t / delta_x * (q[i_t-1, :-1] - q[i_t-1, 1:]), (0, 1), 'edge')

    colors = plt.cm.jet(np.arange(0, 1, 1/len_t))
    for i_t in range(len_t):
        plt.plot(x, q[i_t], color=colors[i_t])
    
    plt.xlabel('x')
    plt.grid()
        
```

```python
len_x = 100
delta_x = 1
len_t = 200
delta_t = 0.01
c = -20
q_init = np.sin(np.arange(0, 5, 5/len_x))

simulation(len_x, delta_x, len_t, delta_t, c, q_init)
```
